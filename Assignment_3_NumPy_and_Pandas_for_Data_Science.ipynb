```python
#Task 1.1: Data Loading and Exploration
import numpy as np
import pandas as pd
# Load the dataset
df = pd.read_csv('StudentsPerformance.csv')

# Your code here:

# 1. Display first 10 rows

print("First 10 rows:")
print(df.head(10))

#2. Show dataset shape and info()

print("\nThe shape of the datasheet is :", df.shape)
print("Dataset Info:")
print(df.info(), "\n")

# 3. Display unique values in each categorical column
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
for col in categorical_cols:
    print(f"Unique values in {col}:", df[col].unique(), "\n")

# 4. Show basic statistics for all score columns
score_cols = ['math score', 'reading score', 'writing score']
print("Basic statistics for score columns:\n", df[score_cols].describe())
```

```python
#Task 1.2: NumPy Array Operations
# Convert score columns to numpy arrays
math_scores = df['math score'].values
reading_scores = df['reading score'].values
writing_scores = df['writing score'].values

# Your tasks:
# 1. Calculate mean, median, std, min, max for each subject using NumPy

subjects = {
    "Math": math_scores,
    "Reading": reading_scores,
    "Writing": writing_scores
}

for subj, scores in subjects.items():
    print(f"{subj} - Mean: {np.mean(scores):.2f}, Median: {np.median(scores):.2f}, "
          f"Std: {np.std(scores):.2f}, Min: {np.min(scores)}, Max: {np.max(scores)}")

# 2. Find total students who scored above 80 in math

math_above_80 = np.sum(math_scores > 80)
print("\nStudents with math > 80:", math_above_80)

# 3. Find total students who scored below 50 in any subject

below_50 = np.sum((math_scores < 50) | (reading_scores < 50) | (writing_scores < 50))
print(f"Students with any score < 50: {below_50}")

# 4. Calculate the overall average score across all three subjects

overall_average = np.mean([np.mean(math_scores), np.mean(reading_scores), np.mean(writing_scores)])
print(f"Overall average score: {overall_average:.2f}")

# Expected output example:
# Math - Mean: 66.1, Median: 66.0, Std: 15.2
# Students with math > 80: 132
# Students with any score < 50: 178

# Your code here:
```

```python
#Task 3.2: Basic Groupby Analysis
# Comprehensive groupby analysis
# Your tasks:

# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students

gender_group = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
gender_count = df.groupby("gender").size()

print("Average Scores by Gender:\n", gender_group)
print("\nCount of Students by Gender:\n", gender_count)

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

test_prep_group = df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()
print("\nAverage Scores by Test Preparation:\n", test_prep_group)

# Find which subject benefits most from test prep
benefit = test_prep_group.loc["completed"] - test_prep_group.loc["none"]
print("\nScore improvement due to Test Preparation:\n", benefit)
print("Most benefited subject:", benefit.idxmax())

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

parent_math = df.groupby("parental level of education")["math score"].mean().sort_values(ascending=False)
print("\nAverage Math Score by Parental Education Level:\n", parent_math)

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

lunch_group = df.groupby("lunch")[["math score", "reading score", "writing score"]].mean()
lunch_count = df.groupby("lunch").size()
print("\nAverage Scores by Lunch Type:\n", lunch_group)
print("\nCount of Students by Lunch Type:\n", lunch_count)

# Your code here:
```

```python
#Task 3.3: Use your data analysis skills to answer these easy questions:

# 1. Who performs better in math - males or females?
#    - Calculate average math score by gender
#    - Show the difference between male and female average

avg_math_by_gender = df.groupby("gender")["math score"].mean()
math_diff = avg_math_by_gender["male"] - avg_math_by_gender["female"]

print("Average Math Score by Gender:\n", avg_math_by_gender)
print("\nDifference (Male - Female):", math_diff)
if math_diff > 0:
    print("Ans: Males perform better in math.")
else:
    print("Ans: Females perform better in math.")

# 2. Which subject do students perform best in overall?
#    - Calculate the overall average for math, reading, and writing
#    - Rank the subjects from highest to lowest average score

overall_avg = df[["math score", "reading score", "writing score"]].mean()
ranked_subjects = overall_avg.sort_values(ascending=False)

print("\nOverall Average Scores by Subject:\n", overall_avg)
print("\nRanked Subjects by Average Score:\n", ranked_subjects)

# 3. What's the impact of parental education?
#    - Find the average total score for each parental education level
#    - Identify which education level leads to highest student performance
df["total score"] = df["math score"] + df["reading score"] + df["writing score"]

parent_total = df.groupby("parental level of education")["total score"].mean().sort_values(ascending=False)

print("\nAverage Total Score by Parental Education Level:\n", parent_total)
print("Ans: Highest student performance is for parents with:", parent_total.idxmax())
```
