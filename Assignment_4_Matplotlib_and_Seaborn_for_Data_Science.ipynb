```python
# Display last 5 rows
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/drive/MyDrive/datasets/gym_membership.csv')
print(df.tail(5))
```

```python
print(df.head())
print("\nThe shape of the data frame is:", df.shape)  # How many rows and columns
print("\nThe parameters of the data frame are:", df.columns.tolist())  # Column names
```

```python
# Check data types
print(df.dtypes)
```

```python
print("\nThe descriptive statistics of the DataFrame are:\n", df.describe())
```

```python
# 1. Bar chart of gym membership types
gender_count = df['gender'].value_counts()

# Simple bar chart
plt.bar(['Female', 'Male'], gender_count.values, color=['green', 'red'])
plt.title('Gender Count')
plt.grid(False)
#plt.savefig("/content/drive/MyDrive/datasets/barchargendervalue.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 1. Bar chart of gym membership types
plt.figure(figsize=(8, 6))
df['abonoment_type'].value_counts().plot(kind='bar', color=['orange', 'blue'])
plt.title('Bar Chart of Gym Membership Types')
plt.xlabel('Membership Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
#plt.savefig("/content/drive/MyDrive/datasets/barchargendervalue.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
#2. Histogram of time in gym
plt.figure(figsize=(8, 6))
plt.hist(df['avg_time_in_gym'], color='blue', edgecolor='black')
plt.title('Histogram of Time in Gym')
plt.xlabel('Average Time in Gym (minutes)')
plt.ylabel('Frequency')
#plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/datasets/barchargendervalue.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 3. Scatter plot: Age vs visits per week
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['visit_per_week'], color='green')
plt.title('Scatter Plot: Age vs Visits per Week')
plt.xlabel('Age')
plt.ylabel('Visits per Week')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/datasets/barchargendervalue.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='gender', y='Age')
plt.title('Box plot of Age vs gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/datasets/barchargendervalue.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 2. Count plot: Days per week
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='visit_per_week')
plt.title('Count Plot of Days per Week')
plt.xlabel('Days per Week')
plt.ylabel('Count')
#plt.xticks(rotation=45, ha='right')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.show()
```

```python
# 3. Bar plot: Average visits by membership type
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='abonoment_type', y='visit_per_week', color ='green')
plt.title('Bar plot of Average Visits by Membership Type')
plt.xlabel('Membership Type')
plt.ylabel('Average Visits per Week')
plt.ylabel('Average Visits per Week')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
```
